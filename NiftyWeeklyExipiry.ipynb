## Finding the day for each entry, So thrusday to thrusday analysis can be done

```python
nifty_data["day"] = nifty_data.apply(lambda row: findDay(row.Date),axis=1)
```

```python

# Python program to Find day of  
# the week for a given date 
import datetime 
import calendar 
  
def findDay(date): 
    born = datetime.datetime.strptime(date, '%Y-%m-%d').weekday() 
    return (calendar.day_name[born]) 
  
# Driver program 
date = '2020-07-23'
print(findDay(date)) 
```

```python

count=0
temp_high = 0
avg_price =0
week_values = []
for index, row in nifty_data.iterrows(): 
    #print(temp_high)
    week_values.append(row["Close"])
        
    if(row["day"] == 'Thursday'):
        avg_price = np.mean(week_values)
        week_avg =week_avg.append({'Date':row["Date"],'Avg':avg_price},ignore_index=True)
        #print("week values-->",week_values)
        #print(temp_high)
        week_values.clear()
   
print(count)
```
